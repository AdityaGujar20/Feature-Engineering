```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA
```

```python
cancer_dataset = load_breast_cancer()
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
df = pd.DataFrame(cancer_dataset['data'], columns = cancer_dataset['feature_names'])
df
```

```python
scalar  = StandardScaler()
scaled_df = pd.DataFrame(scalar.fit_transform(df), columns = df.columns)
scaled_df
```

```python
pca = PCA(n_components = 2)
```

```python
pca_df = pd.DataFrame(pca.fit_transform(scaled_df), columns = ['PC1', 'PC2'])
pca_df
```

```python
sum(pca.explained_variance_)
```

```python
pca2 = PCA()
pca2.fit_transform(scaled_df)
sum(pca2.explained_variance_)
```

```python
plt.figure(figsize = (8, 6))
plt.scatter(x = pca_df['PC1'], y = pca_df['PC2'], c = cancer_dataset['target'])
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Componant 2")
plt.show()
```
